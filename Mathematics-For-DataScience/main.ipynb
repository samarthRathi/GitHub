### User based Function input

```python
from sympy import *

def funcx():
    #error catching loop
    try:
        user_function = input("Enter a mathematical function in terms of 'x': ")
        x = symbols('x')

        allowed_chars = set("0123456789+-*/().x") # setting limits for allowed charachters.
        if not all(c in allowed_chars for c in user_function):
            raise ValueError("Invalid characters in the input")

        # Create the user-defined function using eval
        user_func = eval(user_function)

        # Plotting the function
        plot(user_func)

    except Exception as e:
        print("Error:", e)

funcx()
```

#### Summation using sympy

```python
i,n = symbols('i n')
summation = Sum(2*i, (i,1,n)) # sum formulation goes as (2x1)+(2x2)+(2x3)+(2x4)+(2x5) and the value of i is from 1 to n
n5 = summation.subs(n, 5) # defined n as till 5 i.e., 1,2,3,4,5

#Summations in sympy are lazy, so they do not automatically 
#calculate or get simplified. So we use doit() fun to execute the expression
print(n5.doit())
```

#### Limits using sympy

```python
x = symbols('x')
f = 1/x
res = limit(f, x, oo) #limit of function from x to infinity
print(res)
```

```python
x = symbols('n')
f = (1+(1/n))**n

res = limit(f, n, oo) #limit of function from x to infinity
print(res) # prints it as a euler number : E
print(res.evalf()) # prints the actual value
```

#### Solving Derivatives
* Derivatives in data science quantify how a function changes, crucial for optimizing algorithms, understanding patterns, and enhancing predictive models by evaluating rates of change in datasets.

```python
def derivative_x(f, x, step_size):
    d = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return d

def funcx(x):
    return x**2

slope_2 = derivative_x(funcx, 2, .00001)
print(slope_2)
```

#### Derivative using SymPy

```python
from sympy import * 
x = symbols('x')
f = x**2 # decalring function 

dx_f = diff(f) # calculating the derivative of the function
print(dx_f)
```

```python
def dxf(x):
    return 2*x

slope = dxf(2.0)
print(slope)
```

##### Partial Derivatives and Plotting of the function

```python
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2*x**3 + 3*y**3

#derivative for both the variables x and y
d_fx = diff(f,x)
d_fy = diff(f, y)
print(d_fx)
print(d_fy)
plot3d(f)
```
